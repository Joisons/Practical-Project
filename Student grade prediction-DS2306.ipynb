```python
# Defining the grade mapping dictionary
grade_mapping = {
    'A': 4.0,
    'B': 3.0,
    'C': 2.0,
    'D': 1.0,
    'F': 0.0
}
```

```python
# Converting grades to numerical values in the dataset
for column in df.columns:
    if column != 'CGPA':
        df[column] = df[column].map(grade_mapping)
```
